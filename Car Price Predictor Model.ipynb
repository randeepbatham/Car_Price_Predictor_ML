## Data Cleaning


```python
data_backup = data.copy()
```

```python
data['year'].unique()
```

```python
data = data[data['year'].str.isnumeric()]
```

```python
data['year'] = data['year'].astype(int)
```

```python
data['Price'].str.isnumeric()
```

```python
data.shape
```

```python
data = data[data['Price'] != 'Ask For Price']
```

```python
data['Price'] = data['Price'].str.replace(',','')
```

```python
data['Price'] = data['Price'].astype(int)
```

```python
data.info()
```

```python
data['kms_driven'].unique()
```

```python
data['kms_driven'] = data['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
data = data[data['kms_driven'].str.isnumeric()]
data['kms_driven'] = data['kms_driven'].astype(int)
```

```python
data.info()
```

```python
data = data[~data['fuel_type'].isna()]
```

```python
data['name'] = data['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
data = data.reset_index(drop=True)
```

# Model Building

```python
features = data.drop(columns = 'Price')
target = data['Price']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size = 0.1, random_state = 120)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
encoder = OneHotEncoder(sparse_output = False)
encoder.fit(features[['name','company','fuel_type']])
```

```python
trf_ohe = ColumnTransformer([
    ('ohe',OneHotEncoder(categories = encoder.categories_, sparse_output=False,handle_unknown='ignore'),['name', 'company', 'fuel_type'])
],remainder='passthrough')
```

```python
regression = LinearRegression()
```

```python
pipeline = make_pipeline(trf_ohe, regression)
```

```python
pipeline.fit(xtrain,ytrain)
```

```python
ypred = pipeline.predict(xtest)
```

```python
r2_score(ytest,ypred)
```

```python
scores=[]
for i in range(1000):
    xtrain, xtest, ytrain,ytest = train_test_split(features, target, test_size=0.1,random_state=i)
    regression=LinearRegression()
    pipe=make_pipeline(trf_ohe, regression)
    pipe.fit(xtrain, ytrain)
    ypred=pipe.predict(xtest)
    scores.append(r2_score(ytest,ypred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
xtrain, xtest, ytrain,ytest = train_test_split(features, target, test_size=0.1,random_state=np.argmax(scores))
regression=LinearRegression()
pipe=make_pipeline(trf_ohe, regression)
pipe.fit(xtrain, ytrain)
ypred=pipe.predict(xtest)
r2_score(ytest, ypred)
```
